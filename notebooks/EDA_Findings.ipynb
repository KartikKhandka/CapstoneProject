# EDA Findings & Key Insights

This notebook summarizes the top 10 takeaways from our exploratory data analysis of the Bluestock Fintech mutual fund datasets. We looked at 40 schemes across 10 fund houses, covering historical NAVs, AUM growth, SIP inflows, and investor demographics from 2022 to 2026.

Here is what the data tells us:

### 1. The 2023 Bull Run Lifted Everything
- **What happened:** Around April 2023, almost all 40 schemes started seeing a steady jump in NAV that lasted through the end of the year.
- **Why it matters:** This wasn't just limited to equity funds; the broader market rally pulled up hybrid and even debt funds. It shows how strongly macro sentiment drives overall fund performance.
- **Chart reference:** `reports/charts/01_nav_trend_all_schemes.png`

### 2. SBI Mutual Fund is in a League of Its Own
- **What happened:** Looking at AUM growth from 2022 to 2025, SBI Mutual Fund dwarfs everyone else, crossing ₹12.5 lakh crore by 2025.
- **Why it matters:** Their lead over the second-largest AMC is massive. This comes down to SBI's unmatched banking distribution network and deep reach into tier-2 and tier-3 cities.
- **Chart reference:** `reports/charts/02_aum_growth_fund_house.png`

### 3. SIP Inflows Just Keep Hitting New Records
- **What happened:** Monthly SIP numbers grew consistently month after month, peaking at an all-time high of ₹31,002 crore in December 2025.
- **Why it matters:** Retail investors are treating SIPs like a non-negotiable monthly expense. This constant influx of domestic money gives the market a huge buffer against foreign institutional selling.
- **Chart reference:** `reports/charts/03_sip_inflow_timeseries.png`

### 4. Investors Always Buy the Dip in Equity Funds
- **What happened:** Equity categories saw consistent positive net inflows practically every single month, far outpacing debt and hybrid categories.
- **Why it matters:** Even when the market corrected in late 2024, retail investors didn't panic sell—they actually added more money into equity funds, showing strong confidence in long-term growth.
- **Chart reference:** `reports/charts/04_category_inflow_heatmap.png`

### 5. Mutual Funds are Getting Younger
- **What happened:** The 25–34 age group is now the largest block of active investors in the dataset, closely followed by the 35–44 age bracket.
- **Why it matters:** Fast digital onboarding, UPI, and user-friendly fintech apps have made it incredibly easy for young professionals to start investing early in their careers.
- **Chart references:** `reports/charts/05_age_distribution_pie.png` & `reports/charts/06_sip_boxplot_age.png`

### 6. Big Money is in T30, but Growth is in B30
- **What happened:** Top 30 (T30) cities bring in the bulk of total investment value, but Beyond 30 (B30) cities are catching up fast in terms of new account registrations.
- **Why it matters:** T30 accounts for higher ticket sizes due to higher disposable income. However, B30 is where the next wave of retail adoption is happening, driven by smaller micro-SIPs.
- **Chart reference:** `reports/charts/09_t30_b30_pie.png`

### 7. Folio Counts Have Doubled
- **What happened:** Total industry folios nearly doubled from 13.26 crore in January 2022 to 26.12 crore by December 2025.
- **Why it matters:** Crossing the 20 crore mark is a major milestone for financial inclusion in India. People are moving away from traditional savings accounts and fixed deposits into capital market products.
- **Chart reference:** `reports/charts/10_folio_growth.png`

### 8. Equity Funds Move Together (High Correlation)
- **What happened:** When we look at daily NAV returns across 10 major equity funds, the correlation is highly positive (mostly above 0.85).
- **Why it matters:** Because top active equity funds end up holding many of the same blue-chip stocks, holding five different large-cap funds doesn't actually give you much diversification. To hedge risk, investors need to mix in debt or gold.
- **Chart reference:** `reports/charts/11_return_correlation.png`

### 9. Heavy Bet on Financials and Tech
- **What happened:** Looking at portfolio holdings across equity schemes, the highest sector allocations consistently go to Financial Services and IT.
- **Why it matters:** Since banking and tech make up the heavyweight portion of the NIFTY 50 and SENSEX, active fund managers have to keep heavy allocations here just to track the benchmark.
- **Chart reference:** `reports/charts/12_sector_donut.png`

### 10. Keep an Eye on Expense Ratios
- **What happened:** Comparing expense ratios (TER) against 3-year alpha shows that schemes with lower fees generally have a better shot at delivering positive net alpha.
- **Why it matters:** High management fees eat directly into compounding returns over time. This explains why direct plans with lower TERs have become so popular among informed retail investors.
- **Chart reference:** `reports/charts/17_expense_vs_alpha.png`
